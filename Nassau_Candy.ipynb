```python
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    dayfirst=True
)
print(df["Order Date"].dtype)
```

```python
df['Ship Date'] = pd.to_datetime(
    df['Ship Date'],
    dayfirst= True
)
print(df['Ship Date'].dtype)
```

```python
df['Lead_Time'] = (
    df['Ship Date'] - df['Order Date']
).dt.days

print(df['Lead_Time'].describe())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df['Lead_Time'] = (df['Ship Date'] - df['Order Date']).dt.days

print(df['Lead_Time'].head(20))
```

```python
df[['Order Date','Ship Date','Lead_Time']].head()
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Gross Profit'].sum()
total_units = df['Units'].sum()
avg_lead_time = df['Lead_Time'].mean()
profit_margin= (total_profit/total_sales)*100
```

```python
print("Total_Sales",total_sales)
print("Gross_Profit",total_profit)
print("Total_Units",total_units)
print("Avg_Lead_Time",avg_lead_time)
print("Profit_Margin",profit_margin)
```
